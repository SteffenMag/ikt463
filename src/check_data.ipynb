```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

data = pd.read_csv('../data/data.csv', sep=';')
display(data)
```

```python

display(data['Curricular units 1st sem (approved)'])
# save this column to csv
data['Curricular units 1st sem (approved)'].to_csv('approved1.csv', index=False, header=False)
# plot the data
fig = go.Figure()
fig.add_trace(go.Box(y=data['Curricular units 1st sem (evaluations)'], name='1st sem'))
fig.update_layout(title='1st sem grades')
fig.show()
```

```python
# plot different columns to see the data in histogram, start with first column
# plot all in the same plot with subplots
from plotly.subplots import make_subplots

col = data.columns[-1]
fig = make_subplots(rows=len(data.columns), cols=1)
for i, col in enumerate(data.columns):
    fig.add_trace(go.Histogram(x=data[col], name=col), row=i+1, col=1)
    
fig.update_layout(height=5000, width=800, title_text="Histograms of columns")
fig.show()
```
